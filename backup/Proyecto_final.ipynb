```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns

# Compute the correlation matrix
corr = np.abs(df_tracks.drop(['id', 'name', 'artists', 'id_artists', 'release_date'], axis=1).corr())

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 10))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask,vmin = 0.0, vmax=1.0, center=0.5,
            linewidths=.1, cmap="YlGnBu", cbar_kws={"shrink": .8})

plt.show()
```

**La matriz de correlación muestra las relaciones lineales entre las características del DataFrame df_tracks.**

- **Características más importantes (alta correlación):**
danceability y energy: Alta correlación, sugiriendo que canciones con alta danceability tienden a tener alta energy.
loudness y energy: Alta correlación, indicando que canciones más ruidosas suelen ser más energéticas.
- **Características menos importantes (baja correlación)**:
acousticness y tempo: Baja correlación, sugiriendo que el nivel de acousticness no está relacionado con el tempo.
speechiness y valence: Baja correlación, indicando que estas características aportan información única y no redundante.

```python
import numpy as np
from sklearn.feature_selection import f_regression, mutual_info_regression
import matplotlib.pyplot as plt
import pandas as pd

# Convertimos el DataFrame al formato necesario para scikit-learn
data = df_tracks.drop(columns=['id', 'name', 'artists', 'id_artists', 'release_date'])

# Verificar si hay valores NaN o infinitos y eliminarlos
data = data.replace([np.inf, -np.inf], np.nan).dropna()

# Convertir a valores numpy
data = data.values 

# Nos quedamos con la 1ª columna, popularity, como vector
y = data[:, 0]

# Nos quedamos con el resto
X = data[:, 1:]

# Verificar las dimensiones de X e y
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

# Obtener los nombres de las características
feature_names = df_tracks.drop(columns=['id', 'name', 'artists', 'id_artists', 'release_date']).columns[1:]

# Estudiamos los dos métodos de filtrado
f_test, _ = f_regression(X, y)
f_test /= np.max(f_test)

mi = mutual_info_regression(X, y)
mi /= np.max(mi)

# Visualización
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.bar(range(X.shape[1]), f_test, align="center")
plt.xticks(range(X.shape[1]), feature_names, rotation=90)
plt.xlabel('features')
plt.ylabel('Ranking')
plt.title('$F Test$ score')

plt.subplot(1, 2, 2)
plt.bar(range(X.shape[1]), mi, align="center")
plt.xticks(range(X.shape[1]), feature_names, rotation=90)
plt.xlabel('features')
plt.ylabel('Ranking')
plt.title('Mutual information score')

plt.show()

```

Las gráficas obtenidas a partir del código proporcionado muestran la importancia relativa de cada característica en relación con la variable objetivo (popularity) utilizando dos métodos diferentes de selección de características: el F Test y la Mutual Information.
-Eje X: Muestra los nombres de las características (features) del DataFrame.
-Eje Y: Muestra los valores normalizados del F Test score para cada característica.

**Interpretación de las gráficas:**
Comparación de características: Las características con barras más altas en ambas gráficas son las que tienen una mayor relación con la variable objetivo (popularity). Estas características son potencialmente más importantes para predecir la popularidad de las canciones.
Diferencias entre métodos: Si una característica tiene un alto F Test score pero un bajo Mutual Information score (o viceversa), esto puede indicar que la relación entre esa característica y la variable objetivo es lineal (en el caso del F Test) o no lineal (en el caso de Mutual Information).
