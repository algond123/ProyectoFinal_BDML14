```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import torch

# Cargar dataset
df = final_df_noEncode

# Seleccionar características musicales relevantes
features = ['popularity', 'duration_ms', 'explicit',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
X = df[features]

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convertir a tensor de PyTorch
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
```

```python
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self, input_dim, latent_dim=8):
        super(Autoencoder, self).__init__()
        
        # Codificador
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 16),
            nn.ReLU(),
            nn.Linear(16, latent_dim)
        )
        
        # Decodificador
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 16),
            nn.ReLU(),
            nn.Linear(16, input_dim)
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Definir modelo
input_dim = X_tensor.shape[1]
autoencoder = Autoencoder(input_dim)
```

```python
import torch.optim as optim

# Definir función de pérdida y optimizador
criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.01)

# Entrenamiento
epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    output = autoencoder(X_tensor)
    loss = criterion(output, X_tensor)
    loss.backward()
    optimizer.step()
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
# Obtener la representación latente de las canciones
with torch.no_grad():
    latent_representations = autoencoder.encoder(X_tensor).numpy()
```

```python
from sklearn.cluster import KMeans

# Aplicar K-Means a las representaciones latentes
num_clusters = 6
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(latent_representations)

# Agregar los clusters al DataFrame
df["cluster"] = clusters

# Mostrar distribución de clusters
print(df["cluster"].value_counts())
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reducir dimensiones con PCA
pca = PCA(n_components=2)
latent_2d = pca.fit_transform(latent_representations)

# Graficar Clusters
plt.figure(figsize=(8, 6))
for i in range(num_clusters):
    plt.scatter(latent_2d[clusters == i, 0], latent_2d[clusters == i, 1], label=f"Cluster {i}")

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Clusters de Canciones")
plt.legend()
plt.show()
```

```python
# Agrupar por cluster y calcular la media de las características
cluster_summary = df.groupby("cluster")[features].mean()

# Mostrar las características principales de cada cluster
print("📌 Características principales de cada cluster:")
print(cluster_summary)

import seaborn as sns
import matplotlib.pyplot as plt

# Configurar tamaño del gráfico
plt.figure(figsize=(12, 6))

# Crear un gráfico de barras para comparar características por cluster
sns.heatmap(cluster_summary.T, cmap="coolwarm", annot=True, fmt=".2f")

plt.title("Características Promedio por Cluster")
plt.xlabel("Cluster")
plt.ylabel("Característica")
plt.show()
```
